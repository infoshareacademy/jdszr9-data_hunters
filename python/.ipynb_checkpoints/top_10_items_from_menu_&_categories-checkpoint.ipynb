```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud
```

```python
df = pd.read_csv('restaurant-menus.csv')
```

```python
df
```

# TOP 10 menu categories

```python
df.groupby('category').count().sort_values(['name'], ascending=False).head(10)
```

```python
axCategories = df['category'].value_counts().iloc[:10].plot(kind="barh", color = 'green')
axCategories.invert_yaxis()
axCategories.title.set_text('Top 10 menu categories')
```

```python
plt.figure(figsize=(16,10))
data = df['category'].value_counts().to_dict()
wc = WordCloud(width= 2000, height = 1000, random_state=1,background_color='#191919').generate_from_frequencies(data)
plt.imshow(wc)
plt.title('Top Menu Categories')
plt.axis('off')
plt.show()
```

# ---------------------------------------------

# TOP 10 menu items

```python
df_grouped = df.groupby('name').count().sort_values(['category'], ascending=False).head(10)
df_grouped
```

```python
axMenuItems = df['name'].value_counts().iloc[:10].plot(kind="barh", color = 'green')
axMenuItems.invert_yaxis()
axMenuItems.title.set_text('Top 10 menu items')
```

```python
plt.figure(figsize=(16,10))
data = df['name'].value_counts().to_dict()
wc = WordCloud(width= 2000, height = 1000, random_state=1,background_color='#191919').generate_from_frequencies(data)
plt.imshow(wc)
plt.title('Top Menu Items')
plt.axis('off')
plt.show()
```
